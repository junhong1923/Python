```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1234) #設定隨機種子為１２３４　，　每次跑的結果都達到隨機
data=np.random.randint(-50, 51, 1000) # 範圍-50~50 ， 1000筆資料
# Define sampling frequency and time vector
sf=100. # 頻率100 Hz
time=np.arange(data.size) / sf 

plt.plot(time, data, lw=1, color='b')
plt.xlim([time.min(), time.max()])
plt.title('N3 sleep EEG data(F3)')
plt.xlabel('Time(seconds)')
plt.ylabel('Voltage')
plt.show()
```

# 信號重取樣  
interpolation 增加取樣頻率: 原有訊號太少造成訊號失真  
decimation 降低取樣頻率: 可能造成alias現象(模糊)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate

x=np.arange(0,10.1,0.1)
y=np.exp(-x)
plt.plot(x,y,'b-')


# interpolatin function: interp1d
f=interpolate.interp1d(x,y)

x1=np.arange(0,10.1, 5)
y1=f(x1)
plt.plot(x1, y1, 'ro-')

x2=np.arange(0,10.1, 2)
y2=f(x2)
plt.plot(x2,y2, 'go-')
plt.legend(('Original Signal', 'Resampled Signal', 'Resampled'), loc='best')
plt.show()
```

```python
# interpolatin function: interp1d
f=interpolate.interp1d(x,y)

x1=np.arange(0,10.1, 5)
y1=f(x1)
plt.plot(x1, y1, 'ro-')

x2=np.arange(0,10.1, 2)
y2=f(x2)
plt.plot(x2,y2, 'go-')
plt.legend(('Original Signal', 'Resampled Signal', 'Resampled'), loc='best')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate

x= np.arange(0, 10.1, 2)
# y=np.exp(-x) # 比對1
y=np.cos(-x**2/6.0)

plt.plot(x,y,'b-')
plt.show()
```

```python
f=interpolate.interp1d(x,y, 'quadratic')  #預設'linear'(點與點之間以線性呈現) 'quadratic'
x1=np.arange(0,10.1,0.5)

y1=f(x1)
plt.plot(x1,y1,'ro-')
plt.show()
```
