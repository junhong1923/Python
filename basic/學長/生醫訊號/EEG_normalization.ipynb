```python
import numpy as np
import matplotlib.pyplot as plt

t=np.linspace(-4,4,100)
x=np.sin(t)
print(x)
```

```python
x1=np.sin(t)*1.5 # 放大1.5倍
x2=np.sin(t)*0.5 # 縮小


fig, ax = plt.subplots(1,1, figsize=(8,4))

legend = ['Original Signal', 'Amplify Signal', 'Minify Signal']

plt.plot(x, color = 'k')
plt.plot(x1, color='b')
plt.plot(x2, color='r')

plt.legend(legend)
plt.title('Signal Normalization')
plt.show()
```

```python
fig, ax = plt.subplots(1,1,figsize=(12,4))

plt.hist(x, 50, density= True, facecolor='g', alpha=0.5)
plt.hist(x1, 50, density=True, facecolor='r', alpha=0.5)
plt.hist(x2, 50, density=True, facecolor='b', alpha=0.5)

legend=['Original Signal', 'Amplify Signal', 'Minify Signal']
plt.xlabel=('Signal Value')
plt.ylabel=('Probability')
plt.legend(legend)
plt.title('Signal Normalization')
plt.show()

#　histogram 數值分布:  展現原始訊號調整過後，都會有改變 
```

```python
# 標準化
mean_x = np.mean(x)
std_x = np.std(x)
x_new = (x-mean_x) / std_x

print(mean_x)
print(std_x)

fig, ax = plt.subplots(1, 1, figsize=(8,4))
plt.plot(x_new)
```

```python
mean_x1=np.mean(x1)
std_x1=np.std(x1)
x1_new=(x1-mean_x1)/std_x1

print(mean_x1)
print(std_x1)

fig, ax = plt.subplots(1,1, figsize=(8,4))
plt.plot(x1_new)
```

```python
mean_x2=np.mean(x2)
std_x2=np.std(x2)
x2_new=(x2-mean_x2)/std_x2

print(mean_x2)
print(std_x2)

fig, ax = plt.subplots(1,1, figsize=(8,4))
plt.plot(x2_new)
```

```python
# 標轉化後 比較三個訊號
fig, ax= plt.subplots(1,1,figsize=(8,4))
legend=['Original Signal', 'Amplify Signal', 'Minify Signal']

plt.plot(x_new, color='k')
plt.plot(x1_new, color='b')
plt.plot(x2_new, color='r')

plt.legend(legend)
plt.title('Signal Normalization')
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12,4))
plt.hist(x_new, 50, density=1, facecolor='g', alpha=0.5)
plt.hist(x1_new, 50, density=1, facecolor='r', alpha=0.5)
plt.hist(x2_new, 50, density=1, facecolor='b', alpha=0.5)

legend=['Original Signal', 'Amplify Signal', 'Minify Signal']
plt.legend(legend)
plt.title('Signal Normalization')
plt.show()
```
