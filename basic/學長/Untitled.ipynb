```python
import numpy as np
import matplotlib.pyplot as plt
import math

f = 50 # hz

dt = 1 / (2 * f)
# dt = 1 / (2 * f)
# dt = 20

def func_sin(deg):
    return math.sin(deg * math.pi / 180)

y = np.vectorize(func_sin)

deg_max = 360*25

deg_ls = np.arange(0, deg_max, 1)
# deg_ls = np.arange(0, deg_max, dt)
sin_ls = y(deg_ls)

deg2_ls = np.arange(0, deg_max, 630)
# deg2_ls = np.arange(0, deg_max, 75000*dt)
sin2_ls = y(deg2_ls)

plt.figure()
plt.plot(deg_ls, sin_ls, 'b', deg2_ls, sin2_ls, 'ro')
# plt.plot(deg_ls, sin_ls, 'b', deg2_ls, sin2_ls, 'r--')



```
