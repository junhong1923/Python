#### P.S. 請將所有需要的內容都寫出來


```python
import math
import numpy as np

print(math.radians(30))

def func_deg2rad(deg):
    return deg*math.pi/180
def func_rad2deg(rad):
    return rad*180/math.pi
    
deg = 180
print('deg: %d = rad: %1.2f' %(deg, func_deg2rad(deg)))

rad = 2*math.pi
print('rad: %1.2f = deg: %d' %(rad, func_rad2deg(rad)))
    
```

# Q2 tan 的定義為 sin(theta) / cos(theta) 還是 cos(theta) / sin(theta)？

```python
deg = 30

sin = math.sin(func_deg2rad(deg))
cos = math.cos(func_deg2rad(deg))

tan = math.tan(func_deg2rad(deg))

gauss_1 = sin/cos
gauss_2 = cos/sin


print('sin(%d) = %1.2f' %(deg, sin))
print('cos(%d) = %1.2f' %(deg, cos))
print('tan(%d) = %1.2f' %(deg, tan))

# print('Gauss 1: sin/cos = %1.2f = tan(%d)? : %s' %(gauss_1, deg, gauss_1==tan))
print('Gauss 1: sin/cos = %1.2f = tan(%d)? : %s' %(gauss_1, deg, np.round(gauss_1, 4)==np.round(tan, 4)))

print('Gauss 2: cos/sin = %1.2f = tan(%d)? : %s' %(gauss_2, deg, np.round(gauss_2, 4)==np.round(tan, 4)))
```

# Q6. 比較利用 rand 及 randn 所建立之隨機變數陣列的平均值 (mean) 及標準差 # std)，以迴圈的方式依序比較變數數量 n = 10, 50, 100, 200 之結果, 並以 # print(''" %())的指令印出結果。
### hint: n = [10, 50, 100, 200]


```python
def func_StaticsOfRand(n):
    temp = np.random.rand(n)
    
    return np.mean(temp), np.std(temp)

def func_StaticsOfRandn(n):
    temp = np.random.randn(n)
    
    return np.mean(temp), np.std(temp)

n_ls = [10, 50, 100, 200, 1000, 10000]

for num in n_ls:
    
    rand_m, rand_std = func_StaticsOfRand(num)
    randn_m, randn_std = func_StaticsOfRandn(num)
    
    print('Number of amount: %d' %num)
    print('Rand: mean(std) = %1.4f(%1.4f)' %(rand_m, rand_std))
    print('Randn: mean(std) = %1.4f(%1.4f)\n' %(randn_m, randn_std))

```

# Q9. 請用 pandas 的功能產生一個內容為 6*8 的隨機變數矩陣，index 為 row_i, columns 為 col_i，其中，index 及 column 須以迴圈+陣列方式產生

```python
# V1
import pandas as pd
import numpy as np
num_r = 6
num_c = 8

row = np.random.rand(num_r)
col = np.random.rand(num_c)

matrix = np.zeros((num_r, num_c))

for row_i in range(num_r):
    for col_i in range(num_c):
        
        matrix[row_i, col_i] = row[row_i]*col[col_i]
        
# matrix

matrix_pd = pd.DataFrame(data=matrix)

matrix_pd.to_csv('sample folder/proc.csv')
        
matrix_pd
```
